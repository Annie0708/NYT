```python
import pandas as pd
import faiss
import pickle
import gradio as gr
```

```python
df = pd.read_csv('/content/drive/MyDrive/df_klasyfikacja_small.csv')
```

```python
df.sample()
```

```python
titles = df['title'].tolist()
```

```python
len(titles)
```

```python
# remove duplicates and NaN
titles = [word for word in list(set(titles)) if type(word) is str]
```

```python
from sentence_transformers import SentenceTransformer
# initialize sentence transformer model
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
# create sentence embeddings
sentence_embeddings = model.encode(titles, show_progress_bar=True)
sentence_embeddings.shape
```

```python
#Store sentences & embeddings on disc
with open('embeddings_small.pkl', "wb") as fOut:
    pickle.dump({'sentences': titles, 'embeddings': sentence_embeddings}, fOut, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
#Load sentences & embeddings from disc
with open('/content/drive/MyDrive/embeddings_small.pkl', "rb") as fIn:
    stored_data = pickle.load(fIn)
    stored_sentences = stored_data['sentences']
    sentence_embeddings = stored_data['embeddings']
```

```python
def model_gradio(Title):
  k = 1
  xq = model.encode([Title])
  D, I = index.search(xq, k) 
  a = df['title'].iloc[I.item()]
  return a
```

```python
def model_gradio(Title, Number):
  k = int(Number)
  xq = model.encode([Title])
  D, I = index.search(xq, k) 
  lista=[]
  for i in range(k):
    lista.append(I.item(i))
  a = pd.DataFrame(df['title'].iloc[lista])
  return a
```

```python
demo = gr.Interface(
    fn=model_gradio,
    inputs=[
       'text', 'number'
       ],
    outputs=['numpy'],
    title='Similar articles',
    description='Enter the title and number of similar articles you want to find'
)

demo.launch()
```
